```python
init = True
my_list = []
for c in data:
    if init == True:
        init = False
    else:
        temp_list = data[c].values
        for i in range(len(temp_list)):
            if temp_list[i] == '-':
                temp_list[i] = 0
        my_list.append(temp_list)
```

```python
my_list = np.array(my_list)
rows = len(my_list)
cols = len(my_list[0])

print(rows)
print(cols)
```
